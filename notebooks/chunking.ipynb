```python
import pandas as pd

# Load the cleaned complaints data
df = pd.read_csv('../data/filtered_complaints.csv')
df.head()
```

```python
import sys
sys.path.append('../src')  

from chunking import chunk_narratives
```

```python
chunked_df = chunk_narratives(
    df,
    text_col='cleaned_narrative',
    chunk_size=400,
    chunk_overlap=100,
    metadata_cols=['Product']  
)

chunked_df.head()
```

```python
print("Total chunks:", len(chunked_df))
print("Chunks per product:")
print(chunked_df['Product'].value_counts())
```

```python
chunked_df.to_csv('../data/chunked_complaints.csv', index=False)
print("Saved chunked data to ../data/chunked_complaints.csv")
```
